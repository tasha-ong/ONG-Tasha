```python
import requests as req
import matplotlib.pyplot as plt
import pandas as pd
import json
import time

data = []

for i in range(0,100):
    resp = req.get("http://api.open-notify.org/iss-now.json")
    text = resp.text
    obj = json.loads(text)  
    
    data.append((obj["timestamp"], obj['iss_position']['longitude'], obj['iss_position']['latitude']))
    time.sleep(10)

df = pd.DataFrame(data, columns = ['Timestamp', 'Longitude','Latitude'])
print(pd.DataFrame(df))
```

```python
df = pd.DataFrame(data, columns = ['Timestamp', 'Longitude','Latitude'])
ax1 = df.plot.scatter(x='Latitude',
                      y='Longitude',
                      c='Red',
                      figsize = (30,30)
                      )
ax1.set_title("International Space Station Location")

plt.show()
```
